Question 11 : DÃ©tection par IQR et traitement (2 points)
1. [/1.5] DÃ©tection par Ã©cart interquartile de Quantity :
â Calculez l'IQR (InterQuartile Range) : IQR = Q3 - Q1.
â DÃ©finissez les bornes pour les outliers : limite_inf = Q_1 - 1.5 * IQR et limite_sup =
Q_3 + 1.5 * IQR.
â Comptez le nombre d'outliers dÃ©tectÃ©s par cette mÃ©thode dans la colonne Quantity.
2. [/0.5] Parmi les trois approches suivantes pour traiter les outliers (suppression, imputation
par la mÃ©diane/mode, justification sans modification), laquelle choisiriez-vous et pourquoi ?

```python
# DÃ©tection par Ã©cart interquartile de Quantity :
# â Calculez l'IQR (InterQuartile Range) : IQR = Q3 - Q1.
Q1 = df_sales_clean['Quantity'].quantile(0.25)
Q3 = df_sales_clean['Quantity'].quantile(0.75)
IQR = Q3-Q1

#  DÃ©finissez les bornes pour les outliers : limite_inf = Q_1 - 1.5 * IQR et limite_sup = Q_3 + 1.5 * IQR.
limite_inf = Q1 - 1.5 * IQR
limite_sup = Q3 + 1.5 * IQR

# Comptez le nombre d'outliers dÃ©tectÃ©s par cette mÃ©thode dans la colonne Quantity.
outliers_iqr = df_sales_clean[(df_sales_clean['Quantity'] < limite_inf) | (df_sales_clean['Quantity']) > limite_sup]
print(f"Outliers dÃ©tectÃ©s : {len(outliers_iqr)}")
```

On choisit la justification sans modification parce dans une entreprise commeriale, les grosses commandes sont rÃ©elles et importantes pour le CA. Si on les supprime, on peut fausser les calculs comptables et donc faussÃ©es plus gravement nos analyses.
